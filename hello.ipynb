```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv('New-dataset.csv')
df.head()
```

```python
# Create 'experience' category
def map_experience(years):
    if years <= 0.5:
        return 'fresher'
    elif years <= 1:
        return 'mid-junior'
    elif years <= 1.5:
        return 'junior'
    else:
        return 'mid-senior'

df['Experience'] = df['Years_of_experience'].apply(map_experience)
```

```python
df['Average_of_Project_score'] = (df['Project_1_score']+df['Project_2_score']+df['Project_3_score'])//3
```

```python
df['No_of_certificationsXCGPA']=df['No_of_certifications'] * df['CGPA']
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder1 = OrdinalEncoder(categories=[['beginner', 'fluent', 'expert', 'exceptional']])
encoder2 = OrdinalEncoder(categories=[['fresher', 'mid-junior', 'junior', 'mid-senior' ]])
df['Communication_Level_encoded'] = encoder1.fit_transform(df[['Communication_Level']])
df['Experience_Category_encoded'] = encoder2.fit_transform(df[['Experience']])
```

```python
df.head()
```

```python
X, y = df.drop(columns=['Salary','Communication_Level','Experience']), df['Salary']
X.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

def test_split(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = test_split(X_scaled, y)

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

def print_metrics(y_test, y_pred):
    print("MAE: ", mean_absolute_error(y_test, y_pred))
    print("MSE: ",mean_squared_error(y_test, y_pred))
    print("RMSE: ",np.sqrt(mean_squared_error(y_test, y_pred)))
    print("R2-score: ", r2_score(y_test, y_pred))

print_metrics(y_test, y_pred)
```
